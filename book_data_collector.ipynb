```python
import requests as rq
import pandas as pd
from bs4 import BeautifulSoup
```

```python
web_page = rq.get('https://books.toscrape.com/')
soup = BeautifulSoup(web_page.content,"html.parser")
```

```python
links = soup.select('ol.row article.product_pod a')
links
```

```python
links_list = []
for link in links:
    links_list.append('https://books.toscrape.com/' + link.get('href'))
links_list
```

```python
def get_title(soup_new):
    try: 
        title = soup_new.select_one('div.product_main h1').text.strip()
    except AttributeError:
        title = ''
    return title        
def get_price(soup_new):
    try: 
        price = soup_new.select_one('div.product_main p.price_color').text.strip()
    except AttributeError:
        price = ''
    return price       
def get_pd(soup_new):
    try: 
        pd = soup_new.select_one('#product_description + p').text
    except AttributeError:
        pd = ''
    return pd        
def get_instock(soup_new):
    try: 
        instock = soup_new.select_one('div.product_main p.instock,availability').text.replace("\"",'').strip()
    except AttributeError:
        instock = ''
    return instock        
```

```python
d = {"Title":[],"Price":[],"Product_Description":[],"In_Stock":[]}
for link in links_list:
    new_page = rq.get(link)
    soup_new = BeautifulSoup(new_page.content,"html.parser")
    d['Title'].append(get_title(soup_new))
    d['Price'].append(get_price(soup_new))
    d['Product_Description'].append(get_pd(soup_new))
    d['In_Stock'].append(get_instock(soup_new))
```

```python
d
```

```python
import numpy as np
book_df = pd.DataFrame.from_dict(d)
book_df['Title'].replace('',np.nan,inplace=True)
book_df.dropna(subset=['Title'])
book_df.to_csv('book.csv',header=True,index=False)
```
